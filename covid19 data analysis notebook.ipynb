```python
### FIND MAXIMUM INFECTION RATE FOR ALL COUNTRIES 
## create a new column showing the highest amount of death in a day 'max_infection_rates'
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = [] ## for loop
for c in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated['max_infection_rate'] = max_infection_rates ## note that this tab need be pushed back
```
